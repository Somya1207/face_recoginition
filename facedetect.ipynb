```python
import fitz  # PyMuPDF
import pandas as pd

# Path to the PDF file
pdf_path = 'C:/Users/somya/deepface/Dataset.pdf'

# Open the provided PDF
doc = fitz.open(pdf_path)

# Lists to store data
images = []
names = []
roll_numbers = []

# Iterate through each page
for page_number, page in enumerate(doc):
    # Extract images
    image_list = page.get_images(full=True)
    text = page.get_text("text")
    
    # Parse text to find names and roll numbers (assuming a specific format)
    lines = text.split('\n')
    # print(lines)
    # Temporary storage to hold potential entries
    temp_roll = None
    temp_name = None

    for line in lines:
        # print(line)
        if line.startswith('RA'):  # Typically, roll numbers start with 'RA'
            # print(line)
            parts = line.split()  # Splitting by whitespace
            # print(parts)
            temp_roll = parts[0]  # The roll number is the first part of the split
            # print(parts)
            continue  # Move to the next line to get the name
            
        if temp_roll and (line.isalpha() or ' ' in line):  # Assuming names are alphabetic or common name format
            # print(line)
            temp_name = line
            if 'Not Uploaded' in line or 'Wrong Format' in line:
                # Reset if no valid image is to be associated
                temp_roll = None
                temp_name = None
                continue  # Skip this entry
            
            # Once we have both roll number and name, we look for the image
            for img_index, img in enumerate(image_list):
                xref = img[0]  # xref number
                base_image = doc.extract_image(xref)
                image_bytes = base_image["image"]  # The image in bytes

                if image_bytes:  # Check if image is valid
                    roll_numbers.append(temp_roll)
                    names.append(temp_name)
                    images.append(image_bytes)
                    break  # Stop after associating the first valid image with this entry
            
            # Reset temporary variables after processing an entry
            temp_roll = None
            temp_name = None

# Save to a DataFrame
data = pd.DataFrame({
    'Roll Number': roll_numbers,
    'Name': names,
    'Image': images  # Images are in bytes, you may want to handle them differently
})

# Display the DataFrame
print(data.head())

# Optionally, save the DataFrame to a CSV file
data.to_csv('extracted_data.csv', index=False)
```
